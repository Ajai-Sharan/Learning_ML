```python
from sklearn.datasets import fetch_california_housing
```

```python
dataset = fetch_california_housing()
```

```python
dataset
```

```python
house = pd.DataFrame(data=dataset.data, columns=dataset.feature_names)
```

```python
house
```

```python
house['MedHouseVal'] = dataset.target
house
```

```python
sns.displot(data=house, x="MedHouseVal")
```

```python
sns.distplot(house['MedHouseVal'])
```

## Correlation

 - Positive Correlation.
 - Negative Correlation.

```python
correlation = house.corr()
```

```python
print(correlation)
```

```python
sns.heatmap(correlation)
```

```python
sns.heatmap(correlation, square=True, fmt=".1f", annot=True, annot_kws={'size' : 8}, cmap='Blues')
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(correlation, square=True, fmt=".1f", annot=True, annot_kws={'size' : 8}, cmap='Blues')
```
