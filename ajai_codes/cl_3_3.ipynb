```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
x
```

```python
y = np.sin(x)
y
```

```python
z = np.cos(x)
z
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, z)
plt.show()
```

```python
plt.plot(x, y)
plt.xlabel('Angle')
plt.ylabel('Sine value')
plt.title("Sine wave")
plt.show()
```

```python
x = np.linspace(-10, 10, 20)
y = x**2

plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, '.')
plt.show()
```

```python
plt.plot(x, y, 'r+')
plt.show()
```

```python
plt.plot(x, y, 'c+')
plt.show()
```

```python
x = np.linspace(-5, 5, 50)

plt.plot(x, np.sin(x), 'r')
plt.plot(x, np.cos(x), 'g--')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes( [0, 0, 1, 1] )
languages = ['English', 'French', 'Spanish', 'Latin', 'Tamil']
people = [100, 20, 300, 45, 290]

ax.bar(languages, people)
plt.xlabel("Languages")
plt.ylabel("people")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes( [0, 0, 1, 1] )
languages = ['English', 'French', 'Spanish', 'Latin', 'Tamil']
people = [100, 20, 300, 45, 290]

ax.pie(people, labels=languages, autopct= "%1.1f%%" )
plt.show()
```
